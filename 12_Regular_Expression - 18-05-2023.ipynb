```python
#Write the regular expression and use proper method which gives output as:- 'cat' in str.

import re
str = 'cat mat bat rat'
pattern1 = r"[a-z]{3}"    # Any 3 alphabets 
match1 = re.match(pattern1, str)    # Using match function to get the first 3 characters of the string
print(match1.group(0))    # Using group(0) to extract the exact string from match object
```

```python
#Write the regular expression and use proper method which gives output as:- 'rat' in str.
import re
str = 'cat mat bat rat'
pattern2 = r"[r][a-z]+"    # Searches for a word starting with r and has one or more alphabets
match2 = re.search(pattern2, str)    # Using search function to search throughout the entire string
print(match2.group(0))
```

```python
#Write the regular expression and use proper method which gives output as:- ['mat', 'man'] in str.
import re
str = 'cat mat bat rat man'
pattern3 = r"[m][a-z]+"    # Pattern of a word starting with m and followed by alphabets
match3 = re.findall(pattern3, str)    # To get all the occurances in a list
print(match3)
```

```python
#Write the regular expression and use proper method which gives output as:- 
#['Python', 's', 'Programming', 'is', 'very', 'easy', 'to', 'learn'] in str.
import re
str = "Python's Programming: is very easy to learn"
print(re.findall(r"[a-zA-z]+",str))    # Words with one or more characters
```

```python
#Write the regular expression and use proper method which gives output as:- Python Programming: is very easy to learn in str.
import re
str = "Python's Programming: is very easy to learn"
pattern4 = r"[A-Za-z:]{2,}"    # Words with 2 or more alphabets 
match4 = re.findall(pattern4, str)
out = ""

for i in match4:    # To get the output as a string 
    out = out + ' ' + i

print(out)
```

```python
#Write the regular expression and use proper method which Retrieves all words starting with p.
#Output:- ['peter', 'per', 'picked', 'peck', 'pickled', 'peppers']
import re
str = "peter giper picked a peck of pickled peppers"
pattern5 = r"p[a-z]+"    # Starts with 'p' followed by one or more characters 
match5 = re.findall(pattern5, str)
print(match5)
```

```python
#Write the regular expression and use proper method which Retrieves all words starting with p except 'per' which is not a separate word.
#Output:- ['peter', 'picked', 'peck', 'pickled', 'peppers']
import re
str = "peter giper picked a peck of pickled peppers"
pattern6 = r"\bp[a-z]+\b"    # A word starting with p followed by 2 or more characters within a word boundary 
match6 = re.findall(pattern6, str)
print(match6)
```

```python
#Write the regular expression and use proper method which Retrieves all words starting with a digit.
#Output:- ['8th', '11th']
import re
str = 'The election in delhi will be held on 8th and result for the same will be declared on 11th'
pattern7 = r"\b\d+[a-z]+"    # One or more numbers at the start of the word followed by 1 or more alphabets 
match7 = re.findall(pattern7, str)
print(match7)
```

```python
#Write the regular expression and use proper method which Retrieves last word if starts with c from the given string.
#Output:- ['characters']
import re
str = "Retrieving all words having at least 4 characters"
pattern12 = r"c[a-z]+$"    # Words starting with c if it is the last in given string 
match12 = re.findall(pattern12, str)
print(match12)
```

```python
#Write the regular expression and use proper method which Retrieves a phone number from the given string.
#Output:- 1234567890
import re
str = "Learnbay : 1234567890"
pattern13 = r"\d{10}"    # Exactly 10 digits
match13 = re.search(pattern13, str)
print(match13.group(0))
```

```python
##Write the regular expression and use proper method which Etracts name from the string but not number.
#Output:- Learnbay : 
import re
str = "Learnbay : 1234567890"
pattern14 = r"\D+"    # One or more non digits
match14 = re.search(pattern14, str)
print(match14.group(0))
```
